# Softmax 

Applies the Softmax function to an n-dimensional input Tensor.

Rescales them so that the elements of the n-dimensional output Tensors lie in range [0, 1] and sum 1.

The Softmax function is defined as:

$$
    Softmax(x_i) = \frac{\exp(x_i)}{\sum_{j} \exp(x_j)}
$$

```python
import torch

input = torch.randn(2,3)
print(f"Input\n", input)

output1 = torch.exp(input)/sum(torch.exp(input))
print(f"Output1\n", output1)
output2 = torch.nn.functional.softmax(input)
print(f"Output2\n", output2)
m = torch.nn.Softmax(dim=1)
output3 = m(input)
print(f"Output3\n", output3)
```

# Visulization 

```python
input = torch.linspace(-10,10,steps=100)
softmax = torch.nn.Softmax(dim=0)
output = softmax(input)
output
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 10))
plt.plot(input, output)
plt.title('Softmax Function')
plt.xlabel('Logits')
plt.ylabel('Probability')
plt.grid(True)
plt.legend()
plt.show()
```
